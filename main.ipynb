```python
df.coupon = df.coupon.str.strip()
df.coupon = df.coupon.str.replace('Save ', '')


import re 

def extract_percent(value):
    if '%' in str(value):
        match = re.search(r'([\d.]+)%', value)
        if match: 
            return float(match.group(1))
    return np.nan
df['coupon_discount_percent'] = df['coupon'].apply(extract_percent)
```

```python
def extract_dollar(value):
    if '$' in str(value):
        match = re.search(r'\$([\d.]+)', value)
        if match:
            return float(match.group(1))
    return np.nan
df['coupon_discount_dollar'] = df['coupon'].apply(extract_dollar)
```

```python
def extract_delivery_cost(value):
    if pd.isna(value):
        return np.nan
    
    value = str(value)

    if 'FREE' in value.upper():
        return 0.0

    match = re.search(r'\$([\d.]+)', value)
    if match: 
        return float(match.group(1))

    return np.nan

df['delivery_cost'] = df['delivery_details'].apply(extract_delivery_cost) 
df['delivery_cost'].unique()
```

```python
df['has_fast_delivery'] = df['delivery_details'].str.contains('fastest', case=False, na=False)
df['has_fast_delivery'].unique()
```

```python
df['ships_internationally'] = df['delivery_details'].str.contains('Ships to', case=False, na=False)
```

```python
def extract_delivery_dates(value):
    if pd.isna(value):
        return np.nan, np.nan

    value = str(value)

     # Pattern for date ranges like "Sep 1 - 11" or "Aug 28 - Sep 1"
    range_match = re.search(r'(\w+ \d+)\s*-\s*(\w+ \d+|\d+)', value)
    if range_match: 
        start = range_match.group(1)
        end = range_match.group(2)
        return start, end

    #Pattern for single dates like 'Mon, Sep 1'
    single_match = re.search(r'(\w+,\s*\w+ \d+)', value)
    if single_match:
        date = single_match.group(1)
        return date, date

    return np.nan, np.nan

# Apply and split into two columns
df[['delivery_start', 'delivery_end']] = df['delivery_details'].apply(
    lambda x: pd.Series(extract_delivery_dates(x))
)

    
```
